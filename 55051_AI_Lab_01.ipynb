```python
#Using function check weather given number is even or odd. Take number as input from user

def check_even_odd(number):
    if number % 2 == 0:
        print(f"The number {number} is Even")
    else:
        print(f"The number {number} is Odd")

# Take input from the user
user_input = int(input("Enter a number to check if it's even or odd: "))

# Call the function with user input
check_even_odd(user_input)
```

```python
#Take Best Friend details and print output 5 times using loop

l = input("Enter the name of your best friend:\t\t\t\t")
m = input("What is the age of your best friend:\t\t\t\t")
n = input("What nationality your bestfriend has:\t\t\t\t")
o = input("In what thing your best friend has the most interest:\t\t")

def best_friend():
    print("\n\n<<<<    FOLLOWING ARE THE DETAILS OF YOUR BESTFRIEND YOU SHARED    >>>>\n\n")
    print(f"Name:\t\t\t{l}")
    print(f"Age:\t\t\t{m}")
    print(f"Nationality:\t\t{n}")
    print(f"Interest:\t\t{o}")

for _ in range(5):
    best_friend()
```

```python
#Code to print first N prime numbers starting from a user-defined number

def is_prime(num):
    if num <= 1:
        return False
    for i in range(2, num): # Simple check: iterate up to num-1
        if num % i == 0:
            return False
    return True

# Take starting number and count of primes from user
start_from = int(input("Enter the number from which to start checking for primes: "))
num_of_primes = int(input("Enter how many prime numbers you want to find: "))

prime_numbers = []
count = 0
num = start_from

while count < num_of_primes:
    if is_prime(num):
        prime_numbers.append(num)
        count += 1
    num += 1

print(f"The first {num_of_primes} prime numbers starting from {start_from} are:")
print(prime_numbers)
```
