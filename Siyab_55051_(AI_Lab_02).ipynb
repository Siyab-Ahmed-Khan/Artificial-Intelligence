**PART 1 | Adding Items (The Adders)**

1. append(): Adds an item to the very end

2. insert(): Puts an item at a specific index

3. extend(): Combine two lists together into one big list

```python
# Making a list
phone_list = ["iPhone", "Samsung"]

# Task 01: Adding "Nothing" to the end
phone_list.append("Nothing")

# Task 07: Putting "Redmi" at index 2
phone_list.insert(2, "Redmi")

#Task 05: Add another list of numbers
tablet_list = ["Google tab 2", "iPad 4"]
phone_list.extend(tablet_list)

print(phone_list)
```

**Part 2 | Removing Items (The Cleaners)**
1. remove()
2. pop()
3. clear()

```python
# Lets consider the same list we used above
phone_list = ["iPhone", "Samsung", "Redmi", "Nothing", "Google tab 2", "iPad 4"]

# Task 09: Removing iPhone from list
phone_list.remove("iPhone")
print(phone_list)

# Task 08: Removing the item at index 4
removed_item = phone_list.pop(4)
print(phone_list)

# Task 02: Emptying the whole list
phone_list.clear()
print(phone_list)
```

**Part 3: Searching & Organizing (The Brains)**
1. index(): Tells you the index number of the first item it finds with that name

2. count(): Counts how many times an item appears in the list

3. copy(): Makes a separate "twin" of your list so you don't ruin the original

4. reverse(): Flips the order of the list

5. sort(): Puts items in ABC or 123 order

```python
# Lets consider the same list we used above with little changings in it
phone_list = ["iPhone", "Samsung", "Redmi", "Redmi", "Nothing", "Google tab 2", "iPad 4"]

# Task 06: Finding Nothing phone in list
print(phone_list.index("Nothing"))

# Task 04
print(phone_list.count("Redmi"))

# Task 10 & 11
phone_list.sort()
print(phone_list)

phone_list.reverse()
print(phone_list)

# Task 03: Making a twin
phone_list_copy = phone_list.copy()
print(phone_list_copy)
```

**Bonus Task | Making a GPA Calculator**

```python
# GPA Calculator using Lists, Loop and If-Else

subjects = ["OOP", "DB", "Pak St.", "English"]
credits = [4, 3, 2, 2]

total_points = 0
total_credits = 0

for i in range(len(subjects)):
    marks = int(input("Enter marks for " + subjects[i] + ": "))

    if marks >= 85:
        gp = 4.0
    elif marks >= 75:
        gp = 3.5
    elif marks >= 65:
        gp = 3.0
    elif marks >= 55:
        gp = 2.5
    else:
        gp = 2.0

    total_points += gp * credits[i]
    total_credits += credits[i]

gpa = total_points / total_credits
print("Your GPA is:", round(gpa, 2))
```
